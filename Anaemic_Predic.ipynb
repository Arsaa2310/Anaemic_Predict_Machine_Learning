```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
Anaemic1 = labelencoder.fit_transform(df['Anaemic'])
Anaemic1
```

```python
df['Anaemic_encode']=Anaemic1
```

```python
Sex1 = labelencoder.fit_transform(df['Sex'])
Sex1
```

```python
df.groupby('Sex').count()
```

```python
Sex=[]
for sex in df['Sex']:
    if sex == 'M' or sex=='M ':
        Sex.append(1)
    else:
        Sex.append(0)
len(Sex)
```

```python
df['Sex_encode'] = Sex
df
```

```python
msk = np.random.rand(len(df)) < 0.8
train = df[msk]
test = df[~msk]
```

```python
from sklearn import linear_model

regr = linear_model.LogisticRegression()
x = np.asanyarray(train[['%Red Pixel','%Blue pixel','%Green pixel','Hb','Sex_encode']])
y = np.asanyarray(train[['Anaemic_encode']])
regr.fit (x, y)
print ('Coefficients: ', regr.coef_)
```

```python
y_hat= regr.predict(test[['%Red Pixel','%Blue pixel','%Green pixel','Hb','Sex_encode']])
x = np.asanyarray(test[['%Red Pixel','%Blue pixel','%Green pixel','Hb','Sex_encode']])
y = np.asanyarray(test[['Anaemic_encode']])
print("Mean Squared Error (MSE) : %.2f"
      % np.mean((y_hat - y) ** 2))

print('Variance score: %.2f' % regr.score(x, y))
```

```python
new_data = {
    '%Red Pixel': [43.2555],
    '%Blue Pixel': [25.9025],
    '%Green Pixel': [30.8421],
    'Hb': [6.3],
    'Sex':[1]
}		
new_data_df = pd.DataFrame(new_data)

# Lakukan prediksi dengan model yang sudah dilatih
x_new = np.asanyarray(new_data_df[['%Red Pixel', '%Blue Pixel', '%Green Pixel','Hb','Sex']])
predicted_hb = regr.predict(x_new)

# Hasil prediksi
print('Predicted Hb:', predicted_hb)
```
